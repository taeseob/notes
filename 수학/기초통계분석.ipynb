# Chapter 01. 기초통계 분석
## 1. 통계학이란
### 기술통계학
    - 자료의 특성을 표, 그림, 통계량 등을 사용하여 쉽게 파악할 수 있도록 정리, 요약하는 방법을 다룬다.
### 추론통계학
    - 모집단의 여러 특성에 대해 표본 자료의 정보를 분석하여 추론하는 방법을 다룬다.
### 질적 자료(qualitative data) = 범주형 자료(categorical data)
    - 명목 자료 : 순서의 개념이 없음 (혈액형, 성별, 직업)
    - 순서 자료 : 순서의 개념이 있음 (학점, 선호도, 옷 사이즈)
    - 경우에 따라 질적 자료는 숫자로 표현 가능
### 양적 자료(quantitative data) = 수치형 자료(numerical data)
    - 연속형 자료 : 키, 몸무게, 시간, 길이 온도, 무게 등
    - 이산형 자료 : 교통사고 건수, 고객의 수, 불량품의 수 등
    - 경우에 따라 양적 자료는 범주화가 가능 : 시험 점수를 '수우미양가'로 변환
## 2. 기술통계
### 그래프를 이용한 자료의 정리
    - 질적 자료인 경우
        - 1개 변수 : 막대그림, 파이그림
        - 2개 변수 : 2개 범주에 대한 막대 그림
    - 양적 자료인 경우
        - 1개 변수 : 히스토그램, 상자그림, 줄기-잎그림, 시계열그림
        - 2개 변수 : 산점도
### 측도를 이용한 자료의 정리
    - 분포의 중심에 관한 측도
        - 평균 (mean), 중앙값 (median), 최빈값 (mode)
        - 분포의 모양에 따라 세 측도의 위치가 다름
    - 분포의 퍼진 정도에 관한 측도
        - 분산, 표준편차, 최소/최대값
        - 범위 (range), 백분위수 (quantile), 사분범위 (IQR)
    - 분포의 형태에 관한 측도
        - 왜도 : 분포의 비대칭 정도
        - 첨도 : 분포의 중심이 뾰족한 정도
## 3. 분포함수
### 확률변수와 분포함수
    - 확률변수 (X) : 확률적으로 발생하는 결과에 수치를 부여 한 것
        - 이산형 확률변수 : 주사위를 던져서 나온 숫자, 주사위를 10번 던져서 짝수가 나오는 회수, 병원의 하루 환자 수 등
        - 연속형 확률변수 : 회사의 매출 성장률, 우리나라 연간 강우량 등
    - 확률분포 또는 분포함수 (p, f) : 확률변수의 값과 확률을 대응시키는 관계
### 확률변수가 하나인 경우
    - 확률질량함수 : 이산형 확률변수의 분포함수
$$\begin{array}{c}
p(x_j)=P[X=x_j] \space for \space all \space j=1,2,...\\\
0 \leq p(x_j) \leq 1\\\
\sum_{j=1}^{\infty}p(x_j)=1
\end{array}$$
    - 확률분포의 기대값(평균)과 분산 (이산형의 경우) (예제 : 필기학습자료 p.22)
$$\begin{array}{c}
E[X] = \mu = \sum_{all \space x} xp(x)\\\
V[X] = \sigma^2 = E[(X-\mu)^2] = E(X^2)-\mu^2
\end{array}$$
    - 확률밀도함수 : 연속형 확률변수의 분포함수
$$\begin{array}{c}
\int_a^bf(x)dx = P[a \leq X \leq b] \space for \space all \space -\infty < a < b < \infty \\\
0 \leq \int_a^bf(x)dx \leq 1\\\
\int_{-\infty}^{\infty}f(x)dx = 1
\end{array}$$
    - 누적분포함수 : 음의 무한대부터 특정 값까지 확률밀도함수를 적분한 것. 확률변수가 임의의 구간에 속할 확률을 구할 수 있음.
$$\begin{array}{c}
F(x) = \int_{-\infty}^xf(y)dy\\\
P(X > a) = 1 - F(a)\\\
P(a < X < b) = F(b) - F(a)
\end{array}$$
### 확률변수가 둘 이상인 경우
    - 결합확률분포(joint probability distribution)
$$\begin{array}{c}
p(x,y)=P[X=x,Y=y]\\\
0 \leq p(x,y) \leq 1 \space for \space all \space (x,y)\\\
\sum_{all \space x} \sum_{all \space y} p(x,y) = 1
\end{array}$$
    - 주변확률분포(marginal probability distribution) (예제 : 필기학습자료 p.24)
        X, Y의 결합확률질량함수 p(x,y)로부터 X, Y 각각의 확률질량함수 p(x), p(y)를 유도할 수 있으며 이를 주변확률질량함수라고 한다.
$$\begin{array}{c}
p(x)=P[X=x]=\sum_{all \space y}p(x,y)\\\
p(y)=P[Y=y]=\sum_{all \space x}p(x,y)
\end{array}$$
    - 두 개의 확률변수에 관한 특성값
        공분산 : 두 개의 확률변수 X, Y에 관한 선형적 연관성을 나타내는 특성값
$$\begin{array}{c}
COV[X,Y] = \sigma_{XY} = E[(X-\mu_X)(Y-\mu_Y)] = E[XY] - \mu_X\mu_Y
\end{array}$$

        공분산이 양수이면 X, Y는 양의 선형 관계를 가지고 음수이면 음의 선형 관계를 가진다. 공분산이 가질 수 있는 값의 범위는 X와 Y의 단위에 의존하므로, 선형관계의 강도를 공분산 만으로 파악할 수 없고 자료의 단위를 변경하면 공분산 값도 변한다.
       
        상관계수 : 공분산을 -1 이상 1 이하의 값으로 보정한 지표. 자료의 단위에 의존하지 않음.
$$\begin{array}{c}
CORR[X,Y] = \rho_{XY} = \frac{COV[X,Y]}{S[X]S[Y]}
\end{array}$$

        선형관계의 방향은 공분산과 동일하게 해석할 수 있고, 절대값이 1에 가까울수록 강한 선형관계를 가지는 것으로 해석할 수 있다.
        
        주의사항 : 산점도에서 수평 또는 수직관계를 보이는 경우 선형 관계를 가지는 것처럼 보이지만 공분산은 0이다. 두 확률변수간에 뚜렷한 관계가 있지만 비선형관계인 경우에도 제대로 측정할 수 없다. 공분산 또는 상관계수의 값이 인과관계의 정도를 나타내는것은 아니다.
### 기대값과 분산의 특성
    - 기대값의 특성
$$\begin{array}{c}
E[a] = a\\\
E[aX] = aE[X]\\\
E[X+b] = E[X] + b\\\
E[aX+bY] = aE[X] + bE[Y]
\end{array}$$
    - 분산의 특성
$$\begin{array}{c}
V[a] = 0\\\
V[aX] = a^2 V[X]\\\
V[X+b] = V[X]\\\
V[aX+bY] = a^2 V[X] + b^2 V[Y] + 2ab COV[X,Y]
\end{array}$$
### 이산형 확률분포함수
    - 베르누이분포 : 확률변수의 결과가 1(성공)과 0(실패) 뿐이고, 1이 나올 확률이 p로 일정한 경우 (베르누이 시행)
$$\begin{array}{c}
p(x) = P[X=x] = p^x (1-p)^{1-x},\space \space \space x=0,1\\\
E[X] = p\\\
V[X] = p(1-p)\\\
\end{array}$$
    - 이항분포 : 베르누이 시행을 독립적으로 n번 반복 시행해서 1이 나오는 회수를 확률변수로 정의한 경우 (그래프 : 필기학습자료 p.28)
$$\begin{array}{c}
p(x) = P[X=x] = {n \choose x} p^x (1-p)^{n-x}\\\
E[X] = np\\\
V[X] = np(1-p)\\\
\end{array}$$
### 연속형 확률분포함수
    - 정규분포 :
평균이 $\mu$, 분산이 $\sigma^2$인 경우,
$$\begin{array}{c}
X \sim N(\mu,\sigma^2)\\\
E[X] = \mu\\\
V[X] = \sigma^2\\\
\end{array}$$
    - 표준정규분포 :
$X \sim N(\mu,\sigma^2)$ 일 때, $\frac{X-\mu}{\sigma} \sim N(0,1)$<br>
정규분포를 따르는 임의의 확률변수 $X \sim N(\mu,\sigma^2)$가 어떤 구간에 속할 확률은 표준정규분포의 누적확률분포함수($\Phi$)를 이용하여 표현할 수 있다.
$$\begin{array}{c}
P(X<b) = \Phi \left( \frac{b-\mu}{\sigma} \right) \\\
P(X>a) = 1 - \Phi \left( \frac{a-\mu}{\sigma} \right) \\\
P(a<X<b) = \Phi \left( \frac{b-\mu}{\sigma} \right) - \Phi \left( \frac{a-\mu}{\sigma} \right) \\\
\end{array}$$
z-score of 0.95 = 1.64 (표준정규확률변수의 0.95 분위수) (예제 : 필기학습자료 p.32,33)

    - 카이제곱분포 : 표본분산을 추정하기 위한 확률분포
    - t-분포 : 모집단 한 개를 대상으로, 모분산을 모르는 경우 모평균의 신뢰구간을 추정하기 위한 통계량이 따르는 분포
    - F-분포 : 모집단 두 개를 대상으로, 두 모분산의 비율을 추정하기 위한 분포 (등분산성)
